```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df.head()
```

```python
df.info()
```

<h2>Soal 1: Visualisasi Scatter Plot Perbandingan Kuantitatif</h2>

Pada tugas kali ini kita akan mengamati nilai DewPointAvg (F) dengan mengamati nilai  HumidityAvg (%), TempAvg (F), dan WindAvg (MPG)

Perhatikan bahwa data kita tidaklah siap untuk di analisis, salah satunya tipe data dari DewPointAvg (F), HumidityAvg (%), dan WindAvg (MPG) adalah object, padahalnya data nya ber isi numeric. maka :
- Ubahlah tipe data tersebut menjadi tipe data float

Kemudian: 

- Kalian tidak akan dengan mudah mengubah tipe data tersebut karena column tersebut mempunyai nilai '-' yang dimana tidak bisa di ubah ke bentuk float, maka replace lah terlebih dahulu data yang bernilai '-' dengan nilai NaN, gunakan method .replace(). baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html
- Isi nilai nan dengan nilai sebelumnya di row tersebut. gunakan method .fillna() dengan argument method bernilai 'ffill', baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
- Sekarang ubah tipe datanya dengan float, gunakan method .astype(), baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html

Setelah ini sebagian data siap untuk di jadikan bahan analisis. maka:

Buahlah visualisasi perbandingan kuantitatif scatter plot, sehingga menghasilkan gambar seperti dibawah:

ket: 

- colormap adalah 'coolwarm'
- berikat warna terhadap setiap data poin dengan nilai dari column TempAvgF
- berikan size terhadap setiap data poin dengan nilai dari column WindAvgMPH, kalikan dengan 20 agar size terlihat lebih besar


Berikan pendapat dari insight yang bisa di dapat dari visualisasi perbandingan kuantitatif ini!!!

![alt text](https://drive.google.com/uc?id=1Pp5oIc-39mcw2vrw6xgCRM9GsvyRtPI5)

```python
# Replace '-' to NaN and change NaN to 0
df = df.replace('-', np.nan)
df = df.fillna(method='ffill')

# Replace DewPointAvg (F), HumidityAvg (%), dan WindAvg (MPH) to Numeric(float)
df = df.astype({
    'DewPointAvgF' : 'float64',
    'HumidityAvgPercent' : 'float64',
    'WindAvgMPH' : 'float64'
})
df.info()
```

```python
# Analisis DewPointAvg (F) dengan mengamati nilai HumidityAvg (%), TempAvg (F), dan WindAvg (MPH)

# create canvas
fig, ax = plt.subplots(figsize=(12,8))

# plotting
ax.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], cmap='coolwarm', c=df['TempAvgF'], s=df['WindAvgMPH']*20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 2: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
mean_sales = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
mean_sales
```

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
# create canvas
fig, ax = plt.subplots(figsize=(16,8))
plt.style.use('seaborn')

region = np.arange(4)
action = list(mean_sales.iloc[0])
adventure = list(mean_sales.iloc[1])
fighting = list(mean_sales.iloc[2])
misc = list(mean_sales.iloc[3])
platform = list(mean_sales.iloc[4])
puzzle = list(mean_sales.iloc[5])
racing = list(mean_sales.iloc[6])
role = list(mean_sales.iloc[7])
shooter = list(mean_sales.iloc[8])
simulation = list(mean_sales.iloc[9])
sports = list(mean_sales.iloc[10])
strategy = list(mean_sales.iloc[11])

ax.bar(region-0.42, action, color='#e14a32', width=0.07, label='Action')
ax.bar(region-0.35, adventure, color='#338abd', width=0.07, label='Adventure')
ax.bar(region-0.28, fighting, color='#988dd4', width=0.07, label='Fighting')
ax.bar(region-0.21, misc, color='#777777', width=0.07, label='Misc')
ax.bar(region-0.14, platform, color='#fac15e', width=0.07, label='Platform')
ax.bar(region-0.07, puzzle, color='#8eba42', width=0.07, label='Puzzle')
ax.bar(region, racing, color='#feb5b8', width=0.07, label='Racing')
ax.bar(region+0.07, role, color='#e14a32', width=0.07, label='Role-Playing')
ax.bar(region+0.14, shooter, color='#338abd', width=0.07, label='Shooter')
ax.bar(region+0.21, simulation, color='#988dd4', width=0.07, label='Simulation')
ax.bar(region+0.28, sports, color='#777777', width=0.07, label='Sports')
ax.bar(region+0.35, strategy, color='#fac15e', width=0.07, label='Strategy')

ax.set_ylabel('Mean Sales')
ax.set_xlabel('Region Sales')
ax.set_title('Mean Sales Video Games By Genre')
ax.set_xticklabels([' ',' ', 'NA_Sales', ' ', 'EU_Sales', ' ', 'JP_Sales', ' ', 'Other_Sales'])
plt.legend(frameon=True ,loc='upper right', bbox_to_anchor=(0.5, 0.61, 0.56, 0.5))
plt.show()
```

Jawab:
- Region NA memiliki rata rata penjualan paling tinggi disetiap genre games
- Genre Platform menempati rata rata penjualan tertinggi di Region NA

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
fig, ax = plt.subplots(figsize=(16,8))
plt.style.use('seaborn')

region = np.arange(4)
action = np.array(mean_sales.iloc[0])
adventure = np.array(mean_sales.iloc[1])
fighting = np.array(mean_sales.iloc[2])
misc = np.array(mean_sales.iloc[3])
platform = np.array(mean_sales.iloc[4])
puzzle = np.array(mean_sales.iloc[5])
racing = np.array(mean_sales.iloc[6])
role = np.array(mean_sales.iloc[7])
shooter = np.array(mean_sales.iloc[8])
simulation = np.array(mean_sales.iloc[9])
sports = np.array(mean_sales.iloc[10])
strategy = np.array(mean_sales.iloc[11])

ax.bar(region, action, color='#e14a32', label='Action')
ax.bar(region, adventure, color='#338abd', label='Adventure', bottom=action)
ax.bar(region, fighting, color='#988dd4', label='Fighting', bottom=action+adventure)
ax.bar(region, misc, color='#777777', label='Misc', bottom=action+adventure+fighting)
ax.bar(region, platform, color='#fac15e', label='Platform', bottom=action+adventure+fighting+misc)
ax.bar(region, puzzle, color='#8eba42', label='Puzzle', bottom=action+adventure+fighting+misc+platform)
ax.bar(region, racing, color='#feb5b8', label='Racing', bottom=action+adventure+fighting+misc+platform+puzzle)
ax.bar(region, role, color='#e14a32', label='Role-Playing', bottom=action+adventure+fighting+misc+platform+puzzle+racing)
ax.bar(region, shooter, color='#338abd', label='Shooter', bottom=action+adventure+fighting+misc+platform+puzzle+racing+role)
ax.bar(region, simulation, color='#988dd4', label='Simulation', bottom=action+adventure+fighting+misc+platform+puzzle+racing+role+shooter)
ax.bar(region, sports, color='#777777', label='Sports', bottom=action+adventure+fighting+misc+platform+puzzle+racing+role+shooter+simulation)
ax.bar(region, strategy, color='#fac15e', label='Strategy', bottom=action+adventure+fighting+misc+platform+puzzle+racing+role+shooter+simulation+sports)

ax.set_ylabel('Mean Sales')
ax.set_xlabel('Region Sales')
ax.set_title('Mean Sales Video Games By Genre')
ax.set_xticklabels([' ',' ', 'NA_Sales', ' ', 'EU_Sales', ' ', 'JP_Sales', ' ', 'Other_Sales'])
plt.legend(frameon=True ,loc='upper right', bbox_to_anchor=(0.5, 0.61, 0.56, 0.5))
plt.show()
```

Jawab:
- Jumlah penjualan tertinggi terdapat pada region NA, kedua adalah EU, ketika adalah JP, dan keempat adalah Other
