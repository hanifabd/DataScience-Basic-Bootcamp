Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, cos, label='Cosx', color='r')
ax.plot(x, cos3, label='3 Cosx', color='g', linestyle='--')
ax.plot(x, sin, label='Sin(x-90)', color='b', linestyle='-.')
ax.plot(x, sin2, label='2 Sinx', color='y', linestyle='None', marker='x')
ax.legend(loc='center')
plt.title('Grafik Cos Sin')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.show()
```
