```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(18,8), nrows=1, ncols=3, sharex=True, sharey=True)
ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, cos, color='r', label='sinX')
ax[2].plot(x, sin, color='g', label='cosX')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(14,8))

# Figure Inset Configuration
ax.set_title('Grafik CosX')
# kiri-bawah
insLB = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
insLB.set_title('Grafik CosX^2')
# kiri-atas
insLU = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
insLU.set_title('Grafik CosX^5')
# kanan-bawah
insRB = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
insRB.set_title('Grafik CosX^3')
# kanan-atas
insRU = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
insRU.set_title('Grafik CosX^4')

# plot data to inset figure
# plot cosx
ax.plot(x, y, color='b')
insLB.plot(x, y, color='b')
insLU.plot(x, y, color='b')
insRB.plot(x, y, color='b')
insRU.plot(x, y, color='b')

# plot per figure
insLB.plot(x, y2, color='r')
insLU.plot(x, y5, color='y')
insRB.plot(x, y3, color='g')
insRU.plot(x, y4, color='#000000')

plt.show()
```
