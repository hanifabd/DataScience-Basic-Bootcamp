```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df[df.isna().any(axis=1)]
```

```python
df = df.drop(columns=['Cabin'])
```

```python
# df[df.isna().any(axis=1)]
age_mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)
df['Age'] = df['Age'].astype(int)
df
```

```python
# df[df.isna().any(axis=1)]
# df['Embarked'].value_counts()
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
sex_dict = {
    'female':0,
    'male':1
}
```

```python
df['Sex'] = df['Sex'].replace(sex_dict)
```

```python
df
```

```python
embarked_dict = {
    'S':0,
    'C':1,
    'Q':2
}
```

```python
df['Embarked'] = df['Embarked'].replace(embarked_dict)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

cols = ['Survived','Pclass','Sex', 'SibSp',
       'Parch', 'Embarked']

fig,ax = plt.subplots(math.ceil(len(cols)/3),3,figsize=(20, 12))
ax = ax.flatten()
for a,s in zip(ax,cols):
    sns.countplot(x =s,data = df,ax =a) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
# train
X = df.drop(columns=['PassengerId','Survived', 'Name', 'Ticket', 'Fare'])
y = df['Survived']
```

```python
model = KNeighborsClassifier()
param_grid = {
    'n_neighbors':np.arange(1,20),
    'weights':['distance','uniform']
}
```

```python
gscv = GridSearchCV(model, param_grid, scoring='roc_auc', cv=3, verbose=10)
gscv.fit(X, y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```
