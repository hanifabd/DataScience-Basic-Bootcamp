```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
train = df.copy()
```

```python
train
```

```python
miss_kp = {
    '?':'Wiraswasta'
}

miss_p = {
    '?':'Spesialis'
}


pendidikan = {
    'SD': 1,
    '1st-4th': 2,
    '5th-6th': 3,
    '7th-8th': 4,
    '9th': 5,
    '10th': 6,
    '11th': 7,
    '12th': 8,
    'SMA': 9,
    'Pendidikan Tinggi': 10,
    'D4': 11,
    'D3': 12,
    'Sarjana': 13,
    'Master': 14,
    'Sekolah Professional': 15,
    'Doktor': 16
}

jk = {
    'Perempuan': 0,
    'Laki2': 1
}

gaji = {
    '<=7jt': 0, 
    '>7jt': 1
}
```

```python
train['Pendidikan'] = train['Pendidikan'].replace(pendidikan)
train['Jenis Kelamin'] = train['Jenis Kelamin'].replace(jk)
train['Kelas Pekerja'] = train['Kelas Pekerja'].replace(miss_kp)
train['Pekerjaan'] = train['Pekerjaan'].replace(miss_p)
train['Gaji'] = train['Gaji'].replace(gaji)
```

```python
labels = ["{0} - {1}".format(i, i + 14) for i in range(17, 90, 15)]
train['Umur'] = pd.cut(train['Umur'], range(15, 95, 15), labels=labels)
```

```python
umur_cat = {
    '17 - 31': 1,
    '32 - 46': 2,
    '47 - 61': 3,
    '62 - 76': 4,
    '77 - 91': 5
}
```

```python
train['Umur'] = train['Umur'].replace(umur_cat)
```

```python
train['Umur'].value_counts()
```

```python
trainc = train.copy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

#Using Pearson Correlation
plt.figure(figsize=(12,10))
trainc['Kelas Pekerja']=trainc['Kelas Pekerja'].astype('category').cat.codes
trainc['Pekerjaan']=trainc['Pekerjaan'].astype('category').cat.codes
trainc['Status Perkawinan']=trainc['Status Perkawinan'].astype('category').cat.codes
cor = trainc.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
train_filtered = train[['Umur', 'Pendidikan', 'Status Perkawinan', 'Jam per Minggu', 'Pekerjaan', 'Jmlh Tahun Pendidikan', 'Gaji']]
train_filtered
```

```python
# train_filtered = pd.get_dummies(train_filtered, columns=['Kelas Pekerja'])
train_filtered = pd.get_dummies(train_filtered, columns=['Pekerjaan'])
train_filtered = pd.get_dummies(train_filtered, columns=['Status Perkawinan'])
train_filtered
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X = train_filtered.drop(columns=['Gaji'])
y = train_filtered['Gaji']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=21, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()
ds_train = scaler.fit_transform(X_train)
ds_test = scaler.transform(X_test)
```

```python
cols = X_train.columns
X_train = pd.DataFrame(ds_train, columns=cols)
X_test = pd.DataFrame(ds_test, columns=cols)
```

```python
model = KNeighborsClassifier()
param_grid = {
    'n_neighbors': np.arange(21,50),
    'weights':['uniform','distance']
}

gscv = GridSearchCV(model, param_grid, scoring='roc_auc', cv=3, verbose=10)
gscv.fit(X_train, y_train)
```

```python
print(gscv.best_params_)
print(gscv.best_score_)
```

```python
model = KNeighborsClassifier(n_neighbors=47, weights='uniform')
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test, y_predict))
```
